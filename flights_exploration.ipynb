## Exploratory Data Analysis

### Importing in libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing in the data

```python
data = pd.read_csv('flights-1.csv')
print(data.head())
```

```python

# Boxplot of origin vs delay times
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='dep_delay', data=data)

# Adding labels and title
plt.title('Boxplot of Origin vs Delay Times', fontsize=16)
plt.xlabel('Origin', fontsize=14)
plt.ylabel('Delay Times', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Boxplot of origin vs delay times
plt.figure(figsize=(10, 6))
sns.boxplot(x='origin', y='arr_delay', data=data)

# Adding labels and title
plt.title('Boxplot of Origin vs Delay Times', fontsize=16)
plt.xlabel('Origin', fontsize=14)
plt.ylabel('Delay Times', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Boxplot of origin vs delay times
plt.figure(figsize=(10, 6))
sns.boxplot(x='carrier', y='arr_delay', data=data)

# Adding labels and title
plt.title('Boxplot of Origin vs Delay Times', fontsize=16)
plt.xlabel('Origin', fontsize=14)
plt.ylabel('Delay Times', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter plot of Distance vs Arrival Delay
plt.figure(figsize=(10, 6))
plt.scatter(data['distance'], data['arr_delay'], alpha=0.7, color='blue')

# Adding labels and title
plt.title('Scatterplot of Distance vs Arrival Delay', fontsize=16)
plt.xlabel('Distance', fontsize=14)
plt.ylabel('Arrival Delay', fontsize=14)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Boxplot of origin vs delay times
plt.figure(figsize=(10, 6))
sns.boxplot(x='dest', y='arr_delay', data=data)

# Adding labels and title
plt.title('Boxplot of Origin vs Delay Times', fontsize=16)
plt.xlabel('Origin', fontsize=14)
plt.ylabel('Delay Times', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate the mean arrival delay for each destination
mean_arrival_delay = data.groupby('dest')['arr_delay'].mean()

# Display the result
print(mean_arrival_delay)
```

```python
# Boxplot of origin vs delay times
plt.figure(figsize=(10, 6))
sns.boxplot(x='month', y='dep_delay', data=data)

# Adding labels and title
plt.title('Boxplot of Month vs Delay Times', fontsize=16)
plt.xlabel('Origin', fontsize=14)
plt.ylabel('Delay Times', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Scatter plot of scheduled depart time vs Arrival Delay
plt.figure(figsize=(10, 6))
plt.scatter(data['sched_dep_time'], data['arr_delay'], alpha=0.7, color='blue')

# Adding labels and title
plt.title('Scatterplot of scheduled depart time vs Arrival Delay', fontsize=16)
plt.xlabel('Scheduled Depart time', fontsize=14)
plt.ylabel('Arrival Delay', fontsize=14)

# Show the plot
plt.grid(True)
plt.show()
```

```python
# Convert the 'time_hour' column to datetime format
data['time_hour'] = pd.to_datetime(data['time_hour'])

# Extract the day of the week
data['day_of_week'] = data['time_hour'].dt.day_name()

# Boxplot of day of the week vs arrival delay
plt.figure(figsize=(10, 6))
sns.boxplot(x='day_of_week', y='arr_delay', data=data, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Adding labels and title
plt.title('Boxplot of Day of the Week vs Arrival Delay', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Arrival Delay', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
