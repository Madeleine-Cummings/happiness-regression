## Exploratory Data Analysis

### Importing in libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing in the data

```python
data = pd.read_csv('2019.csv')
print(data.head())
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Economy..GDP.per.Capita.'], data['Happiness.Score'], alpha=0.7, color='blue')

plt.title('Happiness Score vs Economy (GDP per Capita)', fontsize=16)
plt.xlabel('Economy (GDP per Capita)', fontsize=14)
plt.ylabel('Happiness Score', fontsize=14)

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Health..Life.Expectancy.'], data['Happiness.Score'], alpha=0.7, color='blue')

plt.title('Happiness Score vs Economy (GDP per Capita)', fontsize=16)
plt.xlabel('Economy (GDP per Capita)', fontsize=14)
plt.ylabel('Happiness Score', fontsize=14)

plt.grid(True)
plt.show()
```
