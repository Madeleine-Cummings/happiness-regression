```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('pierce_county_house_sales.csv')
print(data.head())
print(data.columns)
```

```python

# Boxplot of sale price and house square feet 
plt.figure(figsize=(10, 6))
plt.scatter(data['attic_finished_square_feet'], data['sale_price'], alpha=0.7, color='blue')
# Adding labels and title
plt.title('Sale price and attic_finished_square_feet ', fontsize=16)
plt.xlabel('attic_finished_square_feet', fontsize=14)
plt.ylabel('Sale p', fontsize=14)

# Show the plot
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
